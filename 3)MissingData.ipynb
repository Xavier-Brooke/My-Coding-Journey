```python
import numpy as np
import pandas as pd
```

**Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing missing Data

```python
df
```

```python
df
```

```python
df.dropna()
```

```python
data2 = {
    'A': [1, 2, np.nan, 4, 5],
    'B': ["MC", 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df2 = pd.DataFrame(data2)
```

```python
df2
```

```python
df2.isna()
```

```python
df2.dropna()
```

```python
df2.dropna(thresh=3)
```

```python
df2.fillna(110)
```
